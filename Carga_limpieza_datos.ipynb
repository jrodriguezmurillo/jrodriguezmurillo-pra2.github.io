```python
popularities = []
for name in playlists_names:
    popularities.append(np.mean(playlists[playlists['playlist_name'] == name]['popularity']))
playlists_names = [i for _,i in sorted(zip(popularities, playlists_names))]
popularities = sorted(popularities)
pd.DataFrame({'playlist':playlists_names[:20],'popularity':popularities[:20]}).to_csv("data/graphics/top_playlists.csv")
```

```python
characteristics = ['danceability', 'energy', 'loudness', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo']
playlists[characteristics] = (playlists[characteristics]-playlists[characteristics].min())/(playlists[characteristics].max()-playlists[characteristics].min())
```

### Weekly Charts
https://www.kaggle.com/datasets/jfreyberg/spotify-chart-data

```python
weekly_charts = pd.read_csv("data/weekly_charts.csv")
weekly_charts = weekly_charts.sort_values(["streams", "date", "position"], ascending=False)
weekly_charts.artists = weekly_charts.artists.apply(literal_eval)
weekly_charts.artists = [' & '.join(i) for i in weekly_charts.artists.tolist()]
weekly_charts.artist_genres = weekly_charts.artist_genres.apply(literal_eval)
weekly_charts.artist_genres = [' & '.join(i) for i in weekly_charts.artist_genres.tolist()]
weekly_charts.head()
```

```python
weekly_charts['year'] = [i.split('-')[0] for i in weekly_charts['date']]
weekly_charts['month'] = [i.split('-')[1] for i in weekly_charts['date']]
weekly_charts['day'] = [i.split('-')[2] for i in weekly_charts['date']]
```

#### Top canciones

```python
grouped_songs = weekly_charts[weekly_charts.country == "global"].groupby('name').sum()['streams'].sort_values(ascending=False).head(30)
```

```python
songs = list(grouped_songs.keys())
streams = [i/1000000 for i in grouped_songs.tolist()]
artists = [weekly_charts[weekly_charts['name'] == i].iloc[0]['artists'] for i in songs]
pd.DataFrame({"streams": streams, "artist": artists, "song": songs}).to_csv("data/graphics/top_songs.csv")
```

#### Top canciones por año

```python
songs = list(grouped_songs.keys())
df_dict = {'song':songs}
years = list(set(weekly_charts.year))
for year in tqdm(sorted(years)):
    for song in songs:
        if year in df_dict.keys():
            df_dict[year].append(sum(weekly_charts[(weekly_charts['year'] == year) & (weekly_charts['name'] == song)]['streams'])/1000000)
        else:
            df_dict[year] = [sum(weekly_charts[(weekly_charts['year'] == year) & (weekly_charts['name'] == song)]['streams'])/1000000]
```

```python
artists = [weekly_charts[weekly_charts['name'] == i].iloc[0]['artists'] for i in songs]
df_dict['artist'] = artists
pd.DataFrame(df_dict).to_csv("data/graphics/top_songs_year.csv")
```

### Característica de las canciones más escuchadas de la última década
https://www.kaggle.com/datasets/cnic92/spotify-past-decades-songs-50s10s?select=2010.csvsongs_characteristics

```python
songs_characteristics = pd.read_csv("data/songs_attributes/2010.csv", index_col=False)
songs = songs_characteristics.title.tolist()
artists = songs_characteristics.artist.tolist()
characteristics = ['bpm', 'nrgy', 'dnce', 'dB', 'live', 'val', 'dur', 'acous','spch']
match_col_names = {
    'bpm': 'beats per minute',
    'nrgy': 'energy',
    'dnce': 'danceability',
    'dB': 'loudness',
    'live': 'liveness',
    'val': 'valence',
    'dur': 'duration',
    'acous': 'acousticness',
    'spch': 'speechiness',
    'pop': 'popularity'
}
songs_characteristics[characteristics] = (songs_characteristics[characteristics]-songs_characteristics[characteristics].min())/(songs_characteristics[characteristics].max()-songs_characteristics[characteristics].min())
```

```python
final_songs = []
max_characteristics = []
for song, artist in zip(songs[:15], artists[:15]):
    song_characteristics = list(songs_characteristics[songs_characteristics['title'] == song][characteristics].iloc[0])
    cond_idx = [i for i,j in enumerate(song_characteristics) if j>0.6]
    final_songs += [f"{artist}\n - " + song] * len(cond_idx)
    max_characteristics += [match_col_names[characteristics[i]] for i in cond_idx]
```

```python
pd.DataFrame({'song':final_songs,'characteristic': max_characteristics}).to_csv("data/graphics/top_characteristics_decade.csv")
```
